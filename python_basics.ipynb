2. What is an interpreter in Python?
   - In Python, an interpreter is the program that reads and executes Python code line by line.

3. What are pre-defined keywords in Python?
   - Pre-defined keywords in Python are special, reserved words that have fixed meanings in the language.
   We cannot use them as variable names, function names, or identifiers because Python uses them to define its syntax and structure.

4. Can keywords be used as variable names?
   - No — Python keywords cannot be used as variable names.
     Keywords have special meanings in Python’s syntax.
     Using them as variable names would confuse the interpreter.

5. What is mutability in Python?
   - Mutability in Python refers to whether an object’s value can be changed after it is created.
   Mutable Objects: These can be changed in place without creating a new object.
   Common mutable types: list,dict,set,bytearray.

6. Why are lists mutable, but tuples are immutable?
   - Lists are mutable and tuples are immutable because of how they are         designed and stored in memory in Python. Their behavior reflects different purposes:

   Lists → Meant for change
   Lists are built for situations where items need to be added, removed, or modified.
   So Python stores them in a flexible structure that allows updating elements.
   
   Tuples → Meant for fixed collections
   Tuples represent sequences that should not change: constants
   fixed records (e.g., coordinates, database rows)
   keys in dictionaries (because immutability = hashable)

7. What is the difference between “==” and “is” operators in Python?
   - == checks whether two objects have the same value/content.
     
     is checks whether two variables refer to the exact same object in memory.

8. What are logical operators in Python?
   - Logical operators in Python are used to combine conditional statements and return Boolean values (True or False). They are essential in decision-making and controlling program flow.

9. What is type casting in Python?
   - Type casting in Python is the process of converting a variable from one data type to another.
   Types of type casting: * Implicit type casting
                          * Explicit type casting

10. What is the difference between implicit and explicit type casting?
    - Implicit Type Casting Performed automatically by Python.It is also called type promotion usually occurs in numeric operations.

     Explicit Type Casting Performed manually by the programmer.It uses built-in functions like int(), float(), str(), list(), etc.It required when Python cannot automatically convert the type.          

11. What is the purpose of conditional statements in Python?
    - Conditional statements in Python are used to execute code only if certain conditions are true. They allow programs to make decisions and control the flow of execution based on different situations.

12. How does the elif statement work?
    - The elif statement in Python stands for “else if”. It is used to check multiple conditions sequentially after an initial if statement.

     How elif works : Python evaluates the if condition first.

If the if condition is True, its block runs and the rest (elif/else) is skipped.

If the if condition is False, Python checks the elif condition(s) in order.

The first elif that is True executes its block, and the rest are skipped.

If none of the conditions are True, the else block (if present) executes.

13. What is the difference between for and while loops?
    - for loop : Used when the number of iterations is known or you are        iterating over a sequence (like a list, tuple, string, or range).
     Iterates over each element of the sequence.

     While loop : Used when the number of iterations is unknown, and you want to repeat as long as a condition is True.The loop continues until the condition becomes False.

14. Describe a scenario where a while loop is more suitable than a for loop.?
    - A while loop is more suitable than a for loop when you don’t know in advance how many times you need to repeat a block of code, and the repetition depends on a dynamic condition rather than a fixed sequence.

     Suppose you want to ask a user to enter a password, and you want to keep asking until the user enters the correct password.
     You don’t know how many attempts it will take in advance, so a while loop is perfect.
    
     correct_password = "python123"

     user_input = ""

     while user_input != correct_password:

           user_input = input("Enter your password: ")

     print("Access granted!")
     
Explanation: The loop continues until the user enters the correct password.

             We don’t know beforehand how many attempts the user will take.

             A for loop would be less suitable because it’s designed for fixed iterations.

```python
# Create a list and tuple in Python, and demonstrate how attempting to change an element works differently
# for each.

# Creating a list and a tuple
my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print("Original list:", my_list)
print("Original tuple:", my_tuple)

# Modifying the list (allowed)
my_list[1] = 200
print("\nList after modification:", my_list)

# Attempting to modify the tuple (NOT allowed)
try:
    my_tuple[1] = 200
except TypeError as e:
    print("\nError when trying to modify tuple:")
    print(e)
```

```python
# Write a function to demonstrate the behavior of mutable and immutable arguments.
def modify_values(num, lst):
    print("\nInside function BEFORE modification:")
    print("num =", num)
    print("lst =", lst)

    # Try modifying immutable argument
    num = num + 10        # creates a new integer object

    # Modify mutable argument
    lst.append(100)       # modifies the existing list

    print("\nInside function AFTER modification:")
    print("num =", num)
    print("lst =", lst)


# Immutable argument
number = 5

# Mutable argument
my_list = [1, 2, 3]

print("Before function call:")
print("number =", number)
print("my_list =", my_list)

modify_values(number, my_list)

print("\nAfter function call:")
print("number =", number)     # unchanged
print("my_list =", my_list)   # changed
```

```python
#  Write a Python program to convert user input from string to integer, float, and boolean types.
# Program to convert user input into different data types

user_input = input("Enter a value: ")

print("\nOriginal (string):", user_input)
print("Type:", type(user_input))

# Convert to integer
try:
    int_value = int(user_input)
    print("\nAs Integer:", int_value)
    print("Type:", type(int_value))
except ValueError:
    print("\nCannot convert to Integer!")

# Convert to float
try:
    float_value = float(user_input)
    print("\nAs Float:", float_value)
    print("Type:", type(float_value))
except ValueError:
    print("\nCannot convert to Float!")

# Convert to boolean
# Any non-empty string is considered True except '0', 'false', etc. (custom handling)
boolean_value = user_input.lower() in ["true", "1", "yes", "y"]
print("\nAs Boolean:", boolean_value)
print("Type:", type(boolean_value))
```

```python
#  Write code to demonstrate type casting with list elements.
# List with string elements
str_list = ["10", "20", "30", "40.5", "True"]

print("Original List:", str_list)

# Convert to integers (only where possible)
int_list = []
for item in str_list:
    try:
        int_list.append(int(item))
    except ValueError:
        int_list.append(f"Cannot convert '{item}' to int")

print("\nInteger Casting Result:")
print(int_list)

# Convert to floats
float_list = []
for item in str_list:
    try:
        float_list.append(float(item))
    except ValueError:
        float_list.append(f"Cannot convert '{item}' to float")

print("\nFloat Casting Result:")
print(float_list)

# Convert to booleans
bool_list = []
for item in str_list:
    # Custom rule: treat 'true', '1', 'yes' as True, else False
    bool_list.append(item.lower() in ["true", "1", "yes"])

print("\nBoolean Casting Result:")
print(bool_list)
```

```python
#  Write a program to reverse a string using a while loop.
# Program to reverse a string using a while loop

# Take input from the user
original_str = input("Enter a string: ")

reversed_str = ""
index = len(original_str) - 1  # Start from the last character

while index >= 0:
    reversed_str += original_str[index]
    index -= 1

print("Reversed string:", reversed_str)
```

```python
# Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# Program to calculate factorial using while loop

# Take input from the user
num = int(input("Enter a non-negative integer: "))

if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1
    while i <= num:
        factorial *= i
        i += 1
    print(f"The factorial of {num} is: {factorial}")
```
